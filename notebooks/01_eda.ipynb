# EDA — Sentiment multilingue (FR, EN, ES, AR)

Exécuter d'abord depuis la racine du projet : `python scripts/build_dataset.py`

Puis charger `data/raw/multilingual_sentiment.parquet`.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ROOT = Path("..").resolve()
df = pd.read_parquet(ROOT / "data" / "raw" / "multilingual_sentiment.parquet")
df["n_chars"] = df["text"].str.len()
df.head()
```

```python
pd.crosstab(df["lang"], df["label"], margins=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
sns.countplot(data=df, x="lang", hue="label", hue_order=[0, 1, 2], ax=ax)
ax.set_title("Labels par langue (0=nég, 1=neu, 2=pos)")
plt.show()
```
